```python
def preflop(cards):
  if (cards[0] == "X" and cards[1] == "X"):
    return 0
  cards = [card for card in cards if card != 'X']
  cards = cards[:2]
  suits = [card[1] for card in cards]
  ranks = sorted([card[0] for card in cards], key='23456789TJQKA'.index)

  if (suits[0] == suits[1]):
    suitsIndex = "S"
  else:
    suitsIndex = "O"
  if (ranks[0] == ranks[1]):
    suitsIndex = ""
  return preflop_equity[ranks[0] + ranks[1] + suitsIndex]
```

```python
def poker_chances(cards, folded):
    """
    Функция принимает список из 7 карт, где от 0 до 5 карт скрыты, и возвращает список шансов выпадения каждой комбинации в покере.
    """

    known_cards = [card for card in cards if card != 'X']
    folded = [card for card in folded if card != 'X']
    deck = [f"{rank}{suit}" for rank in "23456789TJQKA" for suit in "CDHS"]
    deck = [v for v in deck if v not in set(folded) & set(deck)]
    unknown_cards = [card for card in cards if card == 'X']
    remaining_cards = [card for card in deck if card not in cards]
    num_remaining_cards = len(remaining_cards)
    num_unknown_cards = len(unknown_cards)
    num_combinations = len(list(combinations(remaining_cards, num_unknown_cards)))
    keys =['2', '3', '4', '5', '6', '7', '8', '9', 'T', 'J', 'Q', 'K', 'A']
    keys.reverse()

    baseChances = {'Royal flush': 0.000002, 'Straight flush': 0.000012, 'Four of a kind': 0.000240, 'Full house': 0.001441, 'Flush': 0.001981, 'Straight': 0.003546, 'Three of a kind': 0.022569, 'Two pair': 0.047539, 'One pair': 0.424010, 'High card': 0.501570}
    chances = {'Royal flush': 0, 'Straight flush': 0, 'Four of a kind': 0, 'Full house': 0, 'Flush': 0, 'Straight': 0, 'Three of a kind': 0, 'Two pair': 0, 'One pair': 0, 'High card': 0}
    
    for unknown_cards in combinations(remaining_cards, num_unknown_cards):
        hand = known_cards + list(unknown_cards)

        suits = [card[1] for card in hand]
        ranks = sorted([card[0] for card in hand], key='23456789TJQKA'.index)
        ranks.reverse()
        ranks_x = [el for el, _ in groupby(ranks)] # Без повторений
        sumChancesStart = sum(chances.values())

        # Проверяем наличие флеша
        flush = False
        for suit in set(suits):
            if suits.count(suit) >= 5:
                flush = True
                break
        
        # Проверяем наличие стрита
        
        straight = False
        for i in range(len(ranks_x)-4):
            if ranks_x[i:i+5] == keys[keys.index(ranks_x[i]):keys.index(ranks_x[i])+5]:
                ranks_set = ranks_x[i:i+5]
                straight = True
                break
        if ranks[-5:len(ranks)] == ['5', '4', '3', '2', 'A']:
          straight = True
        
        # Проверяем наличие стрит-флеша
        if straight and flush:
            if ranks_set == ['A', 'K', 'Q', 'J', 'T']:
                chances['Royal flush'] += 1
            else:
                chances['Straight flush'] += 1
        
        # Проверяем наличие каре
        for rank in set(ranks):
            if ranks.count(rank) == 4:
                chances['Four of a kind'] += 1
                break
        
        # Проверяем наличие фул-хауса
        three_of_a_kind = False
        two_of_a_kind = False
        num = 0
        for rank in set(ranks):
            if ranks.count(rank) == 3:
                three_of_a_kind = True
                num +=1
            elif ranks.count(rank) == 2:
                two_of_a_kind = True
                num +=1
        if three_of_a_kind and num >= 2:
            chances['Full house'] += 1
        
        # Проверяем наличие флеша
        if flush:
            chances['Flush'] += 1
        
        # Проверяем наличие стрита
        if straight:
            chances['Straight'] += 1

        # Проверяем наличие тройки
        for rank in set(ranks):
            if ranks.count(rank) == 3:
                chances['Three of a kind'] += 1
                break
        
        # Проверяем наличие двух пар
        pairs = []
        for rank in set(ranks):
            if ranks.count(rank) == 2:
                pairs.append(rank)
        if len(pairs) >= 2:
            chances['Two pair'] += 1
        
        # Проверяем наличие одной пары
        if len(pairs) == 1:
            chances['One pair'] += 1
        
        # Если ни одна комбинация не выпала, значит, рука является high card
        if sum(chances.values()) == sumChancesStart:
            chances['High card'] += 1

    # Вычисляем шансы выпадения каждой комбинации
    CHindex = 0
    maxChIndex = 0
    maxChanses = ""
    for key in chances:
        chances[key] /= num_combinations
        CHindex += chances[key] / baseChances[key]
        if (chances[key] > maxChIndex):
          maxChanses = key
          maxChIndex = chances[key]


    chances['index'] = CHindex    
    chances['preflop'] = preflop(cards) 
    chances['maxChanses'] = maxChanses 

    return chances
```

```python
def textToCards(text):
  cards = [text[i:i+2] for i in range(0, len(text), 2)] + ['X' for i in range(0, int(7 - len(text)/2))]
  print(str(cards).replace('S', '♠').replace('C', '♣').replace('H', '♡').replace('D', '♢'))
  return cards
```

```python
def suitsToSymbols(text):  
  text = text.replace('S', '♠').replace('C', '♣').replace('H', '♡').replace('D', '♢')
  return text
```

```python
def buildDF(setOfFlops,enemies):  
  combos = ['Royal flush', 'Straight flush', 'Four of a kind', 'Full house', 'Flush', 'Straight', 'Three of a kind', 'Two pair', 'One pair', 'High card', 'index', 'preflop' , 'maxChanses']
  chances_data = pd.DataFrame({'Combinations':combos})
  
  i=0
  for flop in setOfFlops:    
    cards1 = textToCards(flop.upper())
    print(cards1)
    enemyCards1 = ['X', 'X'] + cards1[2:]
    chances1 = poker_chances(cards1, [])
    chancesEnemy1 = poker_chances(enemyCards1, cards1[:2])

    chances_data[suitsToSymbols(cards1[0]+cards1[1])] = chances_data['Combinations'].map(chances1)
    chances_data["XX"+str(i)] = chances_data['Combinations'].map(chancesEnemy1)
    i += 1
  
  return chances_data
```

```python
# функция для поиска наиболее похожего изображения в наборе
def find_best_match(image, images):
    best_match = None
    best_score = 0
    for name, template in images.items():
        result = cv2.matchTemplate(image, template, cv2.TM_CCOEFF_NORMED)
        score = result.max()

        if score > best_score:
            best_match = name
            best_score = score
    return best_match, best_score
```

```python
# функция для распознавания карт на изображении
def recognize_cards(image, isTable):
    if isTable:
      minArea = 1800
      maxArea = 7000
      cutSuit = [35, 55, 4, 20]
      cutRank = [3, 30, 4, 20]
      suit_images = suit_imagesTable
      rank_images = rank_imagesTable
    else:
      minArea = 500
      maxArea = 2000
      cutSuit = [19, 34, 0, 14]
      cutRank = [0, 25, 0, 19]
      suit_images = suit_imagesHand
      rank_images = rank_imagesHand
    # преобразование изображения в оттенки серого
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # бинаризация изображения
    _, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

    # поиск контуров карт на изображении
    contours, _ = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

    # распознавание мастей и рангов карт
    cards = []
    for contour in contours:
        area = cv2.contourArea(contour)
        if area < minArea or area > maxArea:
            continue
        x, y, w, h = cv2.boundingRect(contour)

        # обрезаем изображение карты
        card_image = image[y:y+h, x:x+w]
        
        # распознаём масть и ранг карты

        suit_image = card_image[cutSuit[0]:cutSuit[1], cutSuit[2]:cutSuit[3]]
        rank_image = card_image[cutRank[0]:cutRank[1], cutRank[2]:cutRank[3]]

        suit, suit_score = find_best_match(suit_image, suit_images)
        rank, rank_score = find_best_match(rank_image, rank_images)
        # добавляем карту в список
        cards.append((rank, suit, rank_score, suit_score))

    # сортируем карты по убыванию процента совпадения
    #cards.sort(key=lambda card: card[2] * card[3], reverse=True)
    #разворачиваем массив
    cards.reverse()
    # выводим список карт с процентами совпадения
    table = ""
    for card in cards:
        #print("{} of {} ({}% match)".format(card[0], card[1], int(card[2] * card[3] * 100)))
        if (int(card[2] * card[3] * 100) > 20):
          table += card[0] + card[1]

    return table
```
